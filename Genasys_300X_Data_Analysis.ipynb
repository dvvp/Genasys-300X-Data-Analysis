```python
import pandas as pd
import numpy as np
import os
import xlrd
```

```python
dir = '/content/'

lst = []

for file in os.listdir(dir):
  if file.endswith(".xlsx"):
    lst.append(file.split("] ")[1])

# these are the different tests we have
wo_id = np.unique(lst).tolist()
wo_id
```

# Maximum Output (Wide-Off, Ch1)

```python
created_final_df = False
for file in os.listdir(dir):
  if file.endswith(".xlsx"):
    if ('Maximum Output (Wide-Off, Ch1) - PASSED.xlsx' in file):
      df = pd.ExcelFile(dir + file).parse('FFT Spectrum')
      df = df.tail(df.shape[0] - 3)
      df.rename(columns = {"FFT Spectrum" : "X", "Unnamed: 1" : "Y"}, inplace = True)
      df = df.set_index("X")
      if not created_final_df:
        final_df = pd.DataFrame(df.index)
        final_df = final_df.set_index("X")
        created_final_df = True
      final_df = pd.concat([final_df, df], axis=1)

max = final_df.max(axis = 1)
min = final_df.min(axis = 1)
final_df = final_df.assign(MAX = max, MIN = min)
final_df = final_df.get(["MAX", "MIN"])

max_output_ch1 = final_df
max_output_ch1
```

# SNR (Wide-Off, Ch1)

```python
snr = []

created_final_df = False
for file in os.listdir(dir):
  if file.endswith(".xlsx"):
    if ("SNR (Wide-Off, Ch1) - PASSED.xlsx" in file):
        df = pd.ExcelFile(dir + file).parse("Signal to Noise Ratio")
        df = df.tail(df.shape[0] - 2)
        df.rename(columns = {"Signal to Noise Ratio" : "Channel", "Unnamed: 1" : "Signal to Noise Ratio", "Unnamed: 2" : "Lower Limit"}, inplace = True)
        df = df.drop(columns = "Unnamed: 3")
        df = df.set_index("Channel")
        if not created_final_df:
          final_df = pd.DataFrame({"Channel" : ["Ch1"]})
          final_df = final_df.set_index("Channel")
          created_final_df = True
        snr.append(df.get("Signal to Noise Ratio").max())

final_df = final_df.assign(MAX = [np.max(snr)], MIN = [np.min(snr)])

snr_ch1 = final_df
snr_ch1
```
